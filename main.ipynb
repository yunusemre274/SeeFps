```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score,mean_absolute_error
from sklearn.linear_model import LinearRegression,RidgeCV,LassoCV,ElasticNetCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder,StandardScaler,OrdinalEncoder,LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from lazypredict.Supervised import LazyRegressor
from sklearn.feature_selection import SelectKBest, f_regression, RFE
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
X=df.drop(columns=['fps','gpuname','cpuname'])
y=df['fps']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21)
```

```python

from sklearn.impute import SimpleImputer

# ==========================================
# 1. KOLONLARI GRUPLANDIRMA VE OTOMATİK TESPİT
# ==========================================

# Sayısal kolonları (int ve float) Pandas'a otomatik olarak bulduruyoruz
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()

# Kategorik kolonları görevlerine göre ayırıyoruz
ordinal_cols = ['gamesetting', 'gpumemorytype', 'gpudirectx', 'gpuopencl', 'gpuopengl', 'gpushadermodel', 'gpuvulkan']
onehot_cols = ['gamename', 'gpuarchitecture', 'gpubus_interface', 'cpumultiplierunlocked']
high_card_cols = ['cpuname', 'gpuname']

# Eğer otomatik bulduğumuz sayısal kolonlar arasında, kazara kategorik listemizde olan bir kolon varsa 
# (örneğin cpumultiplierunlocked 1-0 olarak float gelmişse) çakışmayı önlemek için sayısal listeden çıkarıyoruz.
tum_kategorikler = ordinal_cols + onehot_cols + high_card_cols
numeric_cols = [col for col in numeric_cols if col not in tum_kategorikler]

print(f"Tespit Edilen Sayısal Kolon Sayısı: {len(numeric_cols)}")

# ==========================================
# 2. ALT BORU HATLARI (PIPELINES) KURULUMU
# ==========================================

# A. Sayısal Veriler: Boşlukları ortanca (median) ile doldur ve aynı ölçeğe getir
numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# B. Sıralı (Ordinal) Kategorikler: Boşlukları en çok tekrar edenle doldur ve numaralandır
ordinal_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ord', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])

# C. Sırasız (One-Hot) Kategorikler: Boşlukları doldur ve 1/0 kolonlarına ayır
onehot_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# ==========================================
# 3. ANA TRANSFORMATÖR (PREPROCESSOR)
# ==========================================

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_pipeline, numeric_cols),
        ('ordinal', ordinal_pipeline, ordinal_cols),
        ('onehot', onehot_pipeline, onehot_cols)
        
    ],
    remainder='drop' # Bu 4 gruba girmeyen tüm gereksiz veya hatalı kolonları çöpe at
)

X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)
```

```python
print(pd.DataFrame(X_train_processed).head())
```

```python
def check_model_CV(model,threshold=0.3):
    y_train_pred=model.predict(X_train_processed)
    y_test_pred=model.predict(X_test_processed)
    
    train_score=r2_score(y_train,y_train_pred)
    test_score=r2_score(y_test,y_test_pred)
    
    print("Train Score: ",round(train_score,2))
    print("Test Score: ",round(test_score))
    
    if abs(train_score - test_score) > threshold and train_score > test_score:
        print("⚠️ Model is Overfitting")
    elif train_score < 0.6 and test_score < 0.6:
        print("⚠️ Model is Underfitting")
    else:
        print("✅ Model is Fit")
```

```python
from sklearn.model_selection import cross_val_score 
def cross_val(model, X_train_processed, y_train, cv=5):
    scores = cross_val_score(model, X_train_processed, y_train, cv=cv, scoring='r2')
    print(f"Cross-Validation R2 Scores: {scores}")
    print(f"Mean R2 Score: {np.mean(scores):.4f}")
    print(f"Standard Deviation of R2 Scores: {np.std(scores):.4f}")
    
    if np.std(scores) > 0.1:
        print("⚠️ High variance in cross-validation scores, model may be unstable.")
    else:
        print("✅ Cross-validation scores are consistent, model appears stable.")
```

```python
models_to_run={
    'RandomForestRegressor':RandomForestRegressor(random_state=21),
    'DecisionTreeRegressor':DecisionTreeRegressor(random_state=21),
    'LinearRegression':LinearRegression(),
    'RidgeCV':RidgeCV(),
    'LassoCV':LassoCV(),
    'ElasticNetCV':ElasticNetCV(),
    'GradientBoostingRegressor':GradientBoostingRegressor(random_state=21),
    'XGBRegressor':XGBRegressor(random_state=21),
    'SVR':SVR()
}
```

```python
def run_model(model):
    # Model eğitimi ve tahminleri
    model.fit(X_train_processed, y_train)
    y_train_pred = model.predict(X_train_processed)
    y_test_pred = model.predict(X_test_processed)
    
    # Metrikleri hesaplıyoruz
    train_score = r2_score(y_train, y_train_pred)
    test_score = r2_score(y_test, y_test_pred)
    mse = mean_squared_error(y_test, y_test_pred)
    rmse = np.sqrt(mse)
    
    # İstersen bu print'ler içeride kalmaya devam edebilir
    check_model_CV(model)
    cross_val(model, X_train_processed, y_train)

    return {
        "Train Score (R2)": round(train_score, 4),
        "Test Score (R2)": round(test_score, 4),
        "MSE": round(mse, 4),
        "RMSE": round(rmse, 4)
    }
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import cross_val_score
from IPython.display import display # Tablonun şık görünmesi için

# 1. Sonuçları toplayacağımız boş liste
tum_sonuclar = []

print("Models are training...")

# 2. Sessiz Eğitim Döngüsü (İçeride print yok)
for model_name, model in models_to_run.items():
    
    # Eğit ve Tahmin Et
    model.fit(X_train_processed, y_train)
    y_train_pred = model.predict(X_train_processed)
    y_test_pred = model.predict(X_test_processed)
    
    # Skorları Hesapla
    train_score = r2_score(y_train, y_train_pred)
    test_score = r2_score(y_test, y_test_pred)
    mse = mean_squared_error(y_test, y_test_pred)
    rmse = np.sqrt(mse)
    
    # K-Fold Cross Validation Hesapla (5 parçalı)
    cv_scores = cross_val_score(model, X_train_processed, y_train, cv=5, scoring='r2')
    cv_mean = cv_scores.mean()
    cv_std = cv_scores.std()
    
    # 3. Modelin Fit/Overfit durumunu matematiksel olarak bul (Senin mantığınla)
    if abs(train_score - test_score) > 0.3 and train_score > test_score:
        durum = "⚠️ Overfitting"
    elif train_score < 0.6 and test_score < 0.6:
        durum = "⚠️ Underfitting"
    else:
        durum = "✅ Fit"

    # 4. Her şeyi sözlüğe paketle ve listeye at
    tum_sonuclar.append({
        "Model Name": model_name,
        "Status": durum,
        "CV Mean R2": round(cv_mean, 4),
        "Test R2": round(test_score, 4),
        "Train R2": round(train_score, 4),
        "CV Sapması": round(cv_std, 4),
        "RMSE (Error)": round(rmse, 4)
    })

# 5. Listeyi DataFrame'e çevir ve CV skoruna göre en iyiden en kötüye sırala
df_modeller = pd.DataFrame(tum_sonuclar)
df_modeller = df_modeller.sort_values(by="CV Mean R2", ascending=False).reset_index(drop=True)

# Ekranı temizle ve sadece tabloyu göster
from IPython.display import clear_output
clear_output()
display(df_modeller)
```
