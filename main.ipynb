```python
def show_metrics(model,y_pred):
    r2=r2_score(y_test,y_pred)
    print(f"R2 Score: {r2:.4f}")
    mse=mean_squared_error(y_test,y_pred)
    print(f"MSE: {mse:.4f}")
    mae=mean_absolute_error(y_test,y_pred)
    print(f"MAE: {mae:.4f}")
    rmse=np.sqrt(mse)
    print(f"RMSE: {rmse:.4f}")
```

```python
def check_CV(model,threshold=0.3):
    y_train_pred=model.predict(X_train_processed)
    y_test_pred=model.predict(X_test_processed)
    
    train_score=r2_score(y_train,y_train_pred)
    test_score=r2_score(y_test,y_test_pred)
    
    print("Train Score: ",round(train_score,2))
    print("Test Score: ",round(test_score))
    
    if abs(train_score - test_score) > threshold and train_score > test_score:
        print("⚠️ Model is Overfitting")
    elif train_score < 0.6 and test_score < 0.6:
        print("⚠️ Model is Underfitting")
    else:
        print("✅ Model is Fit")
```

```python
def run_model(model):
    model.fit(X_train_processed,y_train)
    y_pred=model.predict(X_test_processed)
    check_CV(model)
    show_metrics(model,y_pred)
```

```python
models_to_run={
    'RandomForestRegressor':RandomForestRegressor(random_state=21),
    'DecisionTreeRegressor':DecisionTreeRegressor(random_state=21),
    'LinearRegression':LinearRegression(),
    'RidgeCV':RidgeCV(),
    'LassoCV':LassoCV(),
    'ElasticNetCV':ElasticNetCV(),
    'GradientBoostingRegressor':GradientBoostingRegressor(random_state=21),
    'XGBRegressor':XGBRegressor(random_state=21),
    'SVR':SVR()
}
```

```python
for model_name,model in models_to_run.items():
    print(f"Model Name: {model_name}")
    run_model(model)
    print("**"*50)
```

```python
params_rfc={
    'n_estimators':[10,20,50,70,100],
    'max_depth':[None,5,10,20,50,100],
    'min_samples_split':[2,5,10]}
params_knn={
    'n_neighbors':[3,5,7,9,10],
    'weights':['uniform','distance'],
    'metric':['euclidean','manhattan']
}
params_ridge={
    'alpha':[0.1,1.0,10.0,100.0],
    'fit_intercept':[True,False]
    }
params_lasso={
    'alpha':[0.1,1.0,10.0,100.0],
    'fit_intercept':[True,False]}
params_elastic={
    'alpha':[0.1,1.0,10.0,100.0],
    'l1_ratio':[0.1,0.5,0.9],
    'fit_intercept':[True,False]}
params_gradient={
    'loss':['squared_error', 'huber', 'absolute_error'],
    'learning_rate':[0.01,0.1,0.2],
    'n_estimators':[10,20,50,100,200],
    'max_depth':[3,5,10,None]
}
```

```python
for model_name,model in models_to_run.items():
    print(f"Tuning hyperparameters for {model_name}")
    if model_name=='RandomForestRegressor':
        param_grid=params_rfc
    elif model_name=='KNeighborsRegressor':
        param_grid=params_knn
    elif model_name=='RidgeCV':
        param_grid=params_ridge
    elif model_name=='LassoCV':
        param_grid=params_lasso
    elif model_name=='ElasticNetCV':
        param_grid=params_elastic
    elif model_name=='GradientBoostingRegressor':
        param_grid=params_gradient
    else:
        print(f"No hyperparameter grid defined for {model_name}. Skipping...")
        continue
    
    grid_search=GridSearchCV(estimator=model,param_grid=param_grid,
                             scoring='r2',cv=5,n_jobs=-1,verbose=1)
    grid_search.fit(X_train_processed,y_train)
    
    print(f"Best hyperparameters for {model_name}: {grid_search.best_params_}")
    best_model=grid_search.best_estimator_
    
    print(f"Evaluating tuned {model_name}:")
    run_model(best_model)
    print("**"*50)
```
